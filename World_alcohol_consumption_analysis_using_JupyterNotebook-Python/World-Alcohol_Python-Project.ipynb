### 3. Find out and display horizontal bar chart of the alcohol consumption details in the year '1987' where WHO region is 'Western Pacific' and country is 'Vietnam'


```python
res= df[df['Year']==1987].where( (df['WHO region']=='Western Pacific') & (df['Country']=='Viet Nam'))
res1=res.dropna()
res1
```

### 4. Display horizontal bar chart for the year 1987

```python
import matplotlib.pyplot as plt
plt.barh('Beverage Types', 'Display Value', data=res1, label="1987", color='red')
plt.xlabel("ALCOHOL CONSUMPTION")
plt.ylabel("WESTERN PACIFIC - VIET NAM")
plt.legend()
plt.show()
```
